```python
import pandas as pd
import numpy as np

file_path = r"C:\Users\adity\Downloads\Air Pollutants Dataset of Pune City, India\Air Pollutants Dataset of Pune City, India\MITWPU Kothrud AQMS Dataset_Air Pollutants.xlsx"

# Load WITHOUT header interpretation
df_raw = pd.read_excel(file_path, header=None)

print("Raw shape:", df_raw.shape)
print(df_raw.head(8))
```

```python
# Extract header from row 1
header_row = df_raw.iloc[2].tolist()

# Rename first column properly
header_row[0] = "Timestamp"

print("Extracted header:", header_row)
```

```python
df = df_raw.iloc[5:].copy()
df.columns = header_row
df = df.reset_index(drop=True)

print("After restructuring shape:", df.shape)
print(df.head())
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(" ", "")
df.columns = df.columns.str.replace(".", "")
df.columns = df.columns.str.replace("-", "")

print("Normalized columns:", df.columns.tolist())
```

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"], errors="coerce")
df = df.dropna(subset=["Timestamp"])
```

```python
pollutant_cols = [
    "PM10","PM25","PM1",
    "O3","CO","NO","NO2","NOx","CO2"
]

for col in pollutant_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

print("Missing before drop:")
print(df[pollutant_cols].isna().sum())

df = df.dropna().reset_index(drop=True)

df[pollutant_cols] = df[pollutant_cols].astype(np.float32)

print("Final cleaned shape:", df.shape)
```

```python
df = df.sort_values("Timestamp").reset_index(drop=True)

time_diffs = df["Timestamp"].diff().dropna()

print("Unique intervals:")
print(time_diffs.value_counts())
```

```python
X_raw = df[pollutant_cols].values.astype(np.float32)

print("X_raw shape:", X_raw.shape)
```

```python
print("Final shape:", df.shape)
print("Any NaNs left:", df.isna().sum().sum())
print("Dtypes:")
print(df.dtypes)
```

```python
df.to_csv("clean_air_quality.csv", index=False)
```

```python
import hashlib

def file_hash(filename):
    with open(filename, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

print("CSV MD5:", file_hash("clean_air_quality.csv"))
print("NPY MD5:", file_hash("features_float32.npy"))
```
