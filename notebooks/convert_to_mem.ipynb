```python
import numpy as np

def convert(csv_file, mem_file):
    data = np.loadtxt(csv_file, delimiter=",", dtype=int)

    with open(mem_file, "w") as f:
        for val in data.flatten():
            f.write(f"{val & 0xFFFF:04X}\n")

# Convert all layers
convert(r"\FPGAvsPi\weights\W1.csv", "W1.mem")
convert(r"\FPGAvsPi\weights\W2.csv", "W2.mem")
convert(r"\FPGAvsPi\weights\W3.csv", "W3.mem")

convert(r"\FPGAvsPi\weights\b1.csv", "b1.mem")
convert(r"\FPGAvsPi\weights\b2.csv", "b2.mem")
convert(r"\FPGAvsPi\weights\b3.csv", "b3.mem")

print("All .mem files generated.")
```
